```python
# Basic tokenization
def tokenize(text):
    return text.split()
```

```python
tokenize(text1)
```

```python
# Create a vocabulary
vocab = {
    "the": 0,
    "cat": 1,
    "chased": 2,
    "dog": 3,
    "capital": 4,
    "of": 5,
    "france": 6,
    "is": 7,
    "<unk>": 8,
}

def tokenize2(text1):
    parts = text1.split()
    ids = []

    for part in parts:
        if part in vocab:
            value = vocab[part]
        else:
            value = vocab["<unk>"]

        ids.append(value)

    return ids

token_ids1 =tokenize2(text1)
token_ids1
```

```python
import json

with open("tokenizer.json", "r") as f:
    vocab = json.load(f)

def tokenize2(text1):
    parts = text1.split()
    ids = []

    for part in parts:
        if part in vocab:
            value = vocab[part]
        else:
            value = vocab["<unk>"]

        ids.append(value)

    return ids

token_ids1 =tokenize2(text1)
token_ids1
```

```python
reverse_vocab = {id: part for part, id in vocab.items()}
reverse_vocab
```

```python
def detokenize (ids):
    text = ""
    for id in ids:
         part = reverse_vocab[id] 
         text += part + " "
# Modern tokenizer not added blank space but we can add simple tokenizer
    return text.strip()
# Finally, the extra space at the end of the text is removed.
detokenize(token_ids1)


```

```python
from tokenizers import Tokenizer
from tokenizers.models import BPE
from tokenizers.trainers import BpeTrainer
from tokenizers.pre_tokenizers import Whitespace
```

```python
hf_tokenizer = Tokenizer(BPE())
```

```python
hf_tokenizer.pre_tokenizer= Whitespace()

# It separates words based on spaces
```

```python
trainer = BpeTrainer(vocab_size=64, special_tokens=["<unk>"])

hf_tokenizer.train(["text.txt"], trainer)
```

```python
hf_tokenizer.get_vocab_size(), hf_tokenizer.encode(text1).ids
```
